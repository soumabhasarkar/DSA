```python
def findPath(r,c):
    if r < 0 or c < 0:
        return 0
    if r ==0 and c == 0:
        return 1
    
    p1 = findPath(r-1,c)
    p2 = findPath(r,c-1)
    
    return p1+p2

if __name__ == "__main__":
    test_cases = int(input())
    for c in range(test_cases):
        strings = input()
        M = int(strings.split(' ')[0])
        N = int(strings.split(' ')[1])
        print(findPath(M-1,N-1))
```

```python
def changeColor(mat,p,p_c,c_c):
    if p[0] < 0 or p[1]<0 or p[0] >= len(mat) or p[1] >= len(mat[0]):
        return
    if mat[p[0]][p[1]] != p_c :
        return
    
    mat[p[0]][p[1]] = c_c
    
    changeColor(mat,[p[0],p[1]-1], p_c, c_c)
    changeColor(mat,[p[0],p[1]+1], p_c, c_c)
    changeColor(mat,[p[0]-1,p[1]], p_c, c_c)
    changeColor(mat,[p[0]+1,p[1]], p_c, c_c)
    
if __name__ == "__main__":
    test_cases = int(input())
    for c in range(test_cases):
        strings = input()
        M = int(strings.split(' ')[0])
        N = int(strings.split(' ')[1])
        arr_int = list(map(int, input().strip().split()))
        
        mat = [[arr_int[j*N+i] for i in range(N)] for j in range(M)]
        point_color = input().strip().split()
        p = [int(point_color[0]),int(point_color[1])]
        c_c = point_color[2]
        p_c = mat[p[0]][p[1]]
        changeColor(mat, p, p_c,c_c)
        
        for r in range(len(mat)):
            for c in range(len(mat[0])):
                print(mat[r][c], end=" ")
        
# mat = [[1, 2, 3], [4, 5, 6],[4,4,7],[1, 4, 4], [4, 4, 6]]
# p=[2,1]
# p_c = 4
# c_c = 9
# changeColor(mat,p,p_c,c_c)
# mat
```

```python
def JosephusUtil(n, k):
    
    if len(n) == 1:
        return n[0]
    i = 0
    while i < k-1:
        f = n.pop(0)
        n.append(f)
        i+=1
    n.pop(0)
    
    return JosephusUtil(n,k)

if __name__ == "__main__":
    test_cases = int(input())
    for c in range(test_cases):
        strings = input()
        N = int(strings.split(' ')[0])
        k = int(strings.split(' ')[1])
        n = [i+1 for i in range(N)]
        print(Josephus(n,k))
```
